```python
import pandas as pd
from IPython.display import display
```

```python
raw_df = pd.read_csv('dataset.csv')
display(raw_df)
raw_df.describe()
```

```python
import sweetviz as sv
report = sv.analyze(raw_df)
report.show_html()
```

perfect!!! no missing data


```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(raw_df.diagnosis)

df = raw_df.drop(labels=['diagnosis', 'id'], axis = 1)
display(df)
```

```python
features = df.columns

display(features)
```

```python
#normalize data based on standard deviation and mean value
normalized_df = (df - df.mean()) / df.std()
normalized_df = normalized_df.drop_duplicates(subset=None, keep='first', inplace=False)

display(normalized_df)
# check if any two columns are same
for i in normalized_df.columns:
    for j in normalized_df.columns:
        if i != j and all(normalized_df[i] == normalized_df[j]):
            print("column {0} and column {1} are same".format(i, j))

normalized_report = sv.analyze(normalized_df)
normalized_report.show_html('normalized_report.html')
```

Seems like there is no same column ^_^

```python
import plotly.express as px
import plotly.graph_objects as go

sample_data = normalized_df
visualizable_feature_names = normalized_df.columns
figs = []
fig_hist = []
for i, feature_name in enumerate(visualizable_feature_names):
    fig_hist.append(go.Figure())
    for label in label_encoder.classes_:
        fig_hist[i].add_trace(go.Histogram(x=normalized_df[raw_df['diagnosis'] == label][feature_name], name= label))
    fig_hist[i].update_layout(height=300, width=400, title_text=feature_name)
    fig_hist[i].update_layout(barmode='overlay')
    fig_hist[i].update_traces(opacity=0.5)
    fig_hist[i].show()
```

```python
#split data
from sklearn.model_selection import train_test_split

X = df[features]
y = labels
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, accuracy_score

best_depth, best_tree_accuracy = 1, 0
best_clf = None
tree_f1 = 0
for depth in range(1, 10):
    clf = DecisionTreeClassifier(max_depth=depth) # after testing, max_depth = 4 is the best
    clf = clf.fit(x_train, y_train)

    y_tree_predicted = clf.predict(x_test)
    tree_accuracy = accuracy_score(y_test, y_tree_predicted)
    if tree_accuracy > best_tree_accuracy:
        best_depth, best_tree_accuracy = depth, tree_accuracy
        best_clf = clf
        tree_f1 = f1_score(y_test, y_tree_predicted)

print(f'Decision tree at depth {best_depth} Accuracy score=>{best_tree_accuracy}, f1 = {tree_f1}')
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(max_features='sqrt', n_estimators=300, n_jobs=-1, oob_score = True)
rfc.fit(x_train, y_train)
rfc_pred_train = rfc.predict(x_train)
rfc_f1 = 0
# Evaluating on Training set
y_rfc_predicted = rfc.predict(x_test)
rfc_accuracy = accuracy_score(y_test, y_rfc_predicted)
rfc_f1 = f1_score(y_test, y_rfc_predicted)
print(f'random forest Test Accuracy score=> {rfc_accuracy}, f1 = {rfc_f1}')
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier

best_knn_neighbor, best_knn_accuracy = 1, 0
best_knn = None
knn_f1 = 0

for neighbor in range(1, 50): #find the best neighbor count
    knn = KNeighborsClassifier(n_neighbors=neighbor, n_jobs=-1)
    knn.fit(x_train, y_train)

    y_knn_predicted = knn.predict(x_test)
    knn_accuracy = accuracy_score(y_test, y_knn_predicted)

    if knn_accuracy > best_knn_accuracy:
        best_knn_neighbor, best_knn_accuracy = neighbor, knn_accuracy
        best_knn = knn
        knn_f1 = f1_score(y_test, y_tree_predicted)
# Evaluating on Training set
print(f'KNN at k_neighbor {best_knn_neighbor} Accuracy score=>{best_knn_accuracy}, f1 = {knn_f1}')
```

```python
#combined strategy
#use above 3 models to vote
import numpy as np
from collections import Counter

y_combined = np.array([best_clf.predict(x_test), rfc.predict(x_test), best_knn.predict(x_test)], dtype='str').T

output = []
for value in y_combined:
    output.append(Counter(value).most_common()[0][0])

output = np.array(output).astype(int)
combined_accuracy = accuracy_score(y_test, output)
combined_f1 = f1_score(y_test, output)
print(f'By combined strategy the Accuracy score=>{combined_accuracy}, f1={combined_f1}')
```

```python
#Summary

print(f"By using decision tree, accuracy score = {best_tree_accuracy}, f1_score = {tree_f1}")
print(f"By using random forest, accuracy score = {rfc_accuracy}, f1_score = {rfc_f1}")
print(f"By using knn, accuracy score = {best_knn_accuracy}, f1_score = {knn_f1}")
print(f"By using voting stretegy, accuracy score = {combined_accuracy}, f1_score = {combined_f1}")
```
