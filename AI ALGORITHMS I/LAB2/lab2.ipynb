# Data preprocessing

## General Functions

```python
# This function transform labels to numerics.
def data_mapping(df):
    new_df = df[:]
    new_df['school'] = df['school'].map({"GP":0,"MS":1})
    new_df['sex'] = df['sex'].map({"M":0,"F":1})
    new_df['address'] = df['address'].map({"R":0,"U":1})
    new_df['famsize'] = df['famsize'].map({'GT3':1, 'LE3':0})
    new_df['Pstatus'] = df['Pstatus'].map({"T":1,'A':0})
    new_df['Fjob'] = df['Fjob'].map({'at_home':0, 'health':1,'other':2,'services':3,'teacher':4})
    new_df['Mjob'] = df['Mjob'].map({'at_home':0, 'health':1,'other':2,'services':3,'teacher':4})
    new_df['reason'] = df['reason'].map({'course':1, 'other':2 ,'home':3 ,'reputation':4})
    new_df['guardian'] = df['guardian'].map({'mother':1,'father':2,'other':3})
    for col in ['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']:
        new_df[col] = df[col].map({'yes':1, 'no':0})
    return new_df

```

```python
import matplotlib.pyplot as plot
def draw_distribution_bar_chart(df, title):
    subplot = plot.subplot()
    for index, column_name in enumerate(df.columns):
        hist = subplot.hist(df[column_name], alpha = 0.5, label= column_name)

    subplot.set_ylabel('Scores')
    subplot.set_title(title)

    subplot.legend(loc='best')
    return subplot
```

## Math

```python
mat_df = data_mapping(raw_df_mat)
mat_df.head(5)
```

```python
mat_score_distribution = draw_distribution_bar_chart(mat_df[['G1', 'G2', 'G3']], title='Math Score Distribution')
mat_score_distribution.figure.show()
```

Now we can see these distributions are very similar, so we can try to calculate the correlation coefficient.

```python
print(f"The correlation coefficient between G1 and G3 is {mat_df.G3.corr(mat_df.G1)}.")
print(f"The correlation coefficient between G2 and G3 is {mat_df.G3.corr(mat_df.G2)}.")
print(f"The correlation coefficient between G1 and G2 is {mat_df.G1.corr(mat_df.G2)}.")
```

All these 3 correlations are larger than 80 percent, which means students who get higher score in **G1** and **G2** can also get a higher score in **G3**. So we can regard them as one dependent variable and only consider at **G3**.

Based on the above, we can draw a pie chart to see the partion of each grade, and only consider the G3 score.
Firstly, we divide these scores by following rules:<br><br>
80% - 100% -> A <br>
65% - 79% -> B <br>
50% - 64% -> C <br>
0% - 49% -> F <br>

ref: <a herf="https://durhamcollege.ca/wp-content/uploads/112-academic-grading.pdf">Durham college academic grading</a>

```python
def grade(score):
    if score / 20 >= 0.8:
        return 'A'
    elif score / 20 >= 0.65:
        return 'B'
    elif score / 20 >= 0.5:
        return 'C'
    else:
        return 'F'
```

```python
mat_df['grade'] = mat_df['G3'].apply(grade)
mat_df['grade']
```

```python
mat_pie = plot.pie(mat_df.grade.value_counts().values, 
                    labels = mat_df.grade.value_counts().keys(),
                   autopct = '%1.1f%%',
                   )
plot.title('Math G3 Distribution')
plot.legend(loc='upper right',fontsize=8,borderaxespad=0.3)

plot.show()
```

Similarly, try to find the correlation between other columns and G3 score

```python
columns = mat_df.columns[:-4]
print(columns)
```

```python
mat_correlations = []
for column in columns:
    mat_correlations.append(mat_df.G3.corr(mat_df[column]))

for kv in zip(columns, mat_correlations):
    print(f"correlation with {kv[0]} is {kv[1]}")
```

```python
cor_bar = plot.barh(width=mat_correlations, 
                    y = columns,
                   )
plot.title('Math Score correlationships')
plot.legend(loc='upper right',fontsize=8,borderaxespad=0.3)
plot.show()
```
